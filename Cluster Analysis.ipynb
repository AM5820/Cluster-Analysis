```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(1000,noise=0.05,random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1],color='white',edgecolors='black',s=50,alpha=0.7)
plt.show()
```

## Kmeans Clustering

### Applying Elbow method to find the best number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(x,y)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,10))
plt.plot(wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
y_pred= kmeans.fit_predict(x,y)
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],c='red',edgecolors='black')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],c='blue',edgecolors='black')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],color='cyan',edgecolors='black')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],color='green',edgecolors='black')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],color='orange',edgecolors='black')
plt.title('Clustering using KMeans clustering')
plt.show()
```

## Hierarchical clustering

### Applying Dendrogram to find the best number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Clusters')
plt.ylabel('Eucledian distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_pred2 = hc.fit_predict(x,y)
```

```python
plt.scatter(x[y_pred2==0,0],x[y_pred2==0,1],c='orange',edgecolors='black',s=50,alpha=0.7)
plt.scatter(x[y_pred2==1,0],x[y_pred2==1,1],c='blue',edgecolors='black',s=50,alpha=0.7)
plt.title('Clustering using Hierarchical clustering')
plt.show()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan  = DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_pred3 =dbscan.fit_predict(x,y)
```

```python
print(f'number of clusters: {len(set(y_pred3))}')
```

```python
plt.scatter(x[y_pred3==0,0],x[y_pred3==0,1],c='lightblue',edgecolors='black',s=50,alpha=0.7)
plt.scatter(x[y_pred3==1,0],x[y_pred3==1,1],c='orange',edgecolors='black',s=50,alpha=0.7)
plt.title('DBSCAN')
plt.show()
```
